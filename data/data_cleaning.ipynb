```python
import pandas as pd
```

```python
aforos = pd.read_csv("aforos_movistar_arena.csv", header=None)
```

```python
precios = pd.read_csv("precios_movistar_arena.csv")
```

```python
precios = precios.transpose()
```

```python
new_header = precios.iloc[0]
precios = precios[1:]
precios.columns = new_header
precios.head()
```

```python
# pasamos columnas a int

for column in precios.columns[1:]:
    precios[column] = precios[column].str.replace('(\$|,)', '', regex=True).astype(int)
```

```python
aforos.columns = ["sector", "aforo"]
```

```python
aforos.set_index('sector')
```

```python
# nos quedamos con conciertos en movistar arena

precios = precios.where(precios.Lugar == 'Movistar Arena').dropna()
precios.head()
```

```python
precios.drop('Lugar', axis=1, inplace = True)
```

```python
# se va a considerar una distribucion de media arena con sillas
# esto implica las siguientes columnasS: 

media_arena_sillas = precios.columns.where(precios.loc["A-ha"] > 0).dropna()
```

```python
media_arena_sillas
```

```python
precios_media_arena = precios[media_arena_sillas.tolist()]
```

```python
precios_media_arena
```

```python
precios_media_arena
```

```python
prop_vect = precios_media_arena.loc['A-ha'] / precios_media_arena.loc['A-ha']['S: Platea Alta']
```

```python
prop_vect
```

```python
# rellenamos las entradas faltantes en base a las proporciones de la entrada completa

new_matrix = []
for (index, row) in precios_media_arena.iterrows():
    new_row = []
    for i in range(len(row)):
        if row[i] == 0:
            new_row.append(row['S: Platea Alta'] * prop_vect[i])
        else:
            new_row.append(row[i])
    new_matrix.append(new_row)
precios_media_arena_completos = pd.DataFrame(new_matrix)
```

```python
precios_media_arena_completos.columns = precios_media_arena.columns
precios_media_arena_completos.index = precios_media_arena.index
precios_media_arena_completos
```

```python
# aforos media arena
sectores = media_arena_sillas.tolist()
sectores
```

```python
sectores = sectores[2:]
```

```python
sectores
```

```python
aforos_media_arena = aforos.set_index('sector').loc[sectores]
```

```python
aforos_media_arena
```

```python
# limpieza final

aforos_media_arena = aforos_media_arena.rename(index={
    'S: Diamante': 'diamante',
    'S: Platinum': 'platinum',
    'S: Platea Zafiro': 'platea_zafiro',
    'S: Platea Royal': 'platea_royal',
    'S: Golden': 'golden',
    'S: Platea Alta': 'platea_alta',
    'S: Tribuna': 'tribuna'
})
```

```python
aforos_media_arena
```

```python
precios_media_arena_completos = precios_media_arena_completos.rename(columns={
    'S: Diamante': 'diamante',
    'S: Platinum': 'platinum',
    'S: Platea Zafiro': 'platea_zafiro',
    'S: Platea Royal': 'platea_royal',
    'S: Golden': 'golden',
    'S: Platea Alta': 'platea_alta',
    'S: Tribuna': 'tribuna'
})
```

```python
precios_media_arena_completos
```

```python
precios_media_arena_completos = precios_media_arena_completos.drop(['S: Silla de Ruedas', 'S: +1 Silla de Ruedas'], axis=1)
```

```python
# fila de promedios

precios_media_arena_completos
```

```python
summary_ave_data = precios_media_arena_completos.copy().transpose()
summary_ave_data['precio_promedio'] = summary_ave_data.mean(numeric_only=True, axis=1)
summary_ave_data
```

```python
precios_media_arena = summary_ave_data['precio_promedio']
```

```python
precios_media_arena.index.names = ['sector']
precios_media_arena
```

```python
aforos_media_arena.to_csv("../clean_data/aforos_media_arena.csv")
```
